```python
import matplotlib 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
job = pd.read_csv("Jobs.csv")
people = pd.read_csv("People.csv")
```

```python
job_Unemp = job[["FIPS", "State", "County", 
                            "UnempRate2019", 
                            "UnempRate2018", "UnempRate2017", 
                            "UnempRate2016", "UnempRate2015",
                            "UnempRate2014"]]
job_Unemp = job_Unemp.iloc[1: , :]
job_Unemp.columns = ["id", "State", "County", 
                            "2019", "2018", 
                            "2017", "2016", "2015", 
                            "2014"]
job_Unemp = job_Unemp[job_Unemp['id'].astype(str).str.contains("000")==False]
job_Unemp = job_Unemp[job_Unemp['State']=='WY']
job_Unemp
```

```python
job_Unemp = job_Unemp.melt(['id','State','County'], var_name='Year')
job_Unemp.rename(columns={'value':'RateUnemployed'}, inplace=True)
job_Unemp
```

```python
job_emp = job[["FIPS", "State", "County", 
                            "NumEmployed2014", 
                            "NumEmployed2015", "NumEmployed2016", 
                            "NumEmployed2017", "NumEmployed2018",
                            "NumEmployed2019"]]
job_emp = job_emp.iloc[1: , :]
job_emp.columns = ["id", "State", "County", 
                            "2019", "2018", 
                            "2017", "2016", "2015", 
                            "2014"]
job_emp = job_emp[job_emp['id'].astype(str).str.contains("000")==False]
job_emp = job_emp[job_emp['State']=='WY']
job_emp
```

```python
job_emp = job_emp.melt(['id','State','County'], var_name='Year')
job_emp.rename(columns={'value':'NumEmployed'}, inplace=True)
job_emp
```

```python
job_result = pd.merge(job_emp,job_Unemp, on=["id", "State", "County", "Year"])
job_result
```

```python
people_size_year = people[["FIPS", "State", "County", 
                            "TotalPopEst2014", "TotalPopEst2015", 
                            "TotalPopEst2016", "TotalPopEst2017", 
                            "TotalPopEst2018", "TotalPopEst2019"]]
people_size_year = people_size_year.iloc[1: , :]
people_size_year.columns = ["id", "State", "County", 
                            "2014", "2015", 
                            "2016", "2017", "2018", "2019"]
people_size_year = people_size_year[people_size_year['id'].astype(str).str.contains("000")==False]
people_size_year = people_size_year[people_size_year['State']=='WY']
people_size_year
```

```python
people_size = people_size_year.melt(['id','State','County'], var_name='Year')
people_size.rename(columns={'value':'Population'}, inplace=True)
people_size
```

```python
result =pd.merge(people_size,job_result, on=["id", "State", "County", "Year"])
result
```

```python
countries = ['Wyoming','Albany','Big Horn','Campbell','Carbon',
              'Converse','Crook','Fremont','Goshen','Hot Springs',
              'Johnson','Laramie','Lincoln','Natrona','Niobrara',
              'Park','Platte','Sheridan','Sublette','Sweetwater',
              'Teton','Uinta','Washakie','Weston']

colors = ['#e7ba52', '#a7a7a7', '#aec7e8', '#1f17b4', '#9367bd',
         '#r9ba53', '#b1a7a8', '#kec1e7', '#2f73b3', '#8465cd',
         '#d2ba54', '#c2a7a9', '#bec3e6', '#7f47b2', '#6445bd',
         '#a1ba55', '#d3a7a0', '#fec2e5', '#4f57b1', '#5464ad'
         ]
```

```python
import altair as alt

source = result

scale = alt.Scale(domain=countries,
                  range=colors)
color = alt.Color('County:N', scale=scale)

brush = alt.selection_interval(encodings=['x'])

click = alt.selection_multi(encodings=['color'])

points = alt.Chart().mark_point().encode(
    alt.X('RateUnemployed:O', title='RateUnemployed'),
    alt.Y('Population:Q',
        title='Population',
        scale=alt.Scale(domain=[0, 110000])
    ),
    color=alt.condition(brush, color, alt.value('lightgray')),
    size=alt.Size('NumEmployed:Q', scale=alt.Scale(range=[5, 200]))
).properties(
    width=700,
    height=300
).add_selection(
    brush
).transform_filter(
    click
)
```

```python
bars = alt.Chart().mark_bar().encode(
    x='count()',
    y='County:N',
    color=alt.condition(click, color, alt.value('lightgray')),
).transform_filter(
    brush
).properties(
    width=700,
).add_selection(
    click
)

alt.vconcat(
    points,
    bars,
    data=source,
    title="Population with employeement number and unemployeement rate in WY"
)
```
